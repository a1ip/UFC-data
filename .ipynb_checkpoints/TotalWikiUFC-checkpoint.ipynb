```python
import wikipedia
import csv
from bs4 import BeautifulSoup
import requests
import re
import urllib
```

```python
# Scraping list of UFC events, events: (event name, date)
# must use url title, not event title
# urllib unquote - foramtting for foreign characters
page = wikipedia.WikipediaPage('List of UFC events').html()
html = BeautifulSoup(page, 'html.parser')
table = html.find('table', {'id' : 'Past_events'})
events=[]
for row in table.findAll('tr'):
    try:
        name = urllib.parse.unquote((row.findAll('td')[1].find('a')['href'].split("/")[-1].replace('_',' ')))
        date = row.findAll('td')[2].text.strip()
        events.append((name, date))
    except:
        pass
```

```python
len(events)
```

```python
# scraping fight data
#
datafile = open('allUFCfights.csv', 'w')
writer = csv.writer(datafile, delimiter=',')
writer.writerow(["weight","fighter1","fighter2","method","round_end","time","date","event"])
def scrapeWiki(events):
    page = wikipedia.WikipediaPage(events[0]).html()
    soup = BeautifulSoup(page,'html.parser')
    fights = soup.find('table', {'class':'toccolours'}).findAll('tr')
    for fight in fights:
        cells = fight.findAll('td')
        try:
            weight = cells[0].text
            fighter1 = cells[1].text
            fighter2 = cells[3].text
            method = cells[4].text
            roundend = cells[5].text
            time = cells[6].text
            date = events[1]
            writer.writerow([weight,fighter1,fighter2,method,roundend,time,date,events[0]])
        except:
            pass

errorList=[]
for e in events:
    print("\nTrying: "+e[0])
    try:
        scrapeWiki(e)
    except:
        print("URL read error: "+e[0])
        errorList.append(e[0])
```
