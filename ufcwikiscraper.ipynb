```python
import csv
from bs4 import BeautifulSoup
import requests
import re
```

```python
page='https://en.wikipedia.org/wiki/List_of_UFC_events'
```

```python
r = requests.get(page)
html = BeautifulSoup(r.text, 'html.parser')
table = html.find('table', {'id' : 'Past_events'})
```

```python
urls=[]
for row in table.tbody.findAll('tr'):
    try:
        urls.append(row.findAll('td')[1].find('a').text)
    except:
        pass
```

```python
file = open('ufceventnames.txt', 'w')
for e in urls:
    file.write(e+'\n')
```
