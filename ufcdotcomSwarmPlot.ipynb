```python
import pandas as pd
```

```python
df = pd.read_csv("ufcfightersedit.csv")
```

```python
print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
fig, axes = plt.subplots()
sns.violinplot('Weight','Height', data=df, ax=axes, color="w")
sns.despine(left=True)
axes.set_title('Height distribution for UFC weight classes')
axes.set_xlabel('Weight Class')
axes.set_ylabel('Height (cm)')
plt.ylim(145,225)
plt.figure(figsize=(20,10))
plt.savefig('test.png')
plt.close()
```

```python
sns.set(style="whitegrid")
fig, axes = plt.subplots()
sns.boxplot('Weight','Height', data=df, ax=axes)
axes.set_title('Height distribution for UFC weight classes')
axes.set_xlabel('Weight Class')
axes.set_ylabel('Height (cm)')
plt.ylim(145,225)
plt.xticks
plt.close()
```

```python
sns.set(style="whitegrid")
fig, axes = plt.subplots()
sns.stripplot('Weight','Height', data=df, ax=axes)
axes.set_title('Height distribution for UFC weight classes')
axes.set_xlabel('Weight Class')
axes.set_ylabel('Height (cm)')
plt.ylim(145,225)
plt.xticks
plt.close()
```

```python
sns.set(style="darkgrid")
fig, axes = plt.subplots()
sns.swarmplot('Weight','Height', data=df, ax=axes)
axes.set_title('UFC Height Distribution')
axes.set_xlabel('Weight Class (lbs)')
axes.set_ylabel('Height (cm)')
sns.despine(left=True)
plt.ylim(145,225)
plt.savefig('swarm.png')
```

```python
sns.set(style="whitegrid")
fig, axes = plt.subplots()
sns.boxenplot('Weight','Height', data=df, ax=axes)
axes.set_title('Height distribution for UFC weight classes')
axes.set_xlabel('Weight Class')
axes.set_ylabel('Height (cm)')
plt.ylim(145,225)
plt.close()
```
