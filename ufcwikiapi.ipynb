```python
datafile = open('wikifights.csv', 'w')
writer = csv.writer(datafile, delimiter=',')
writer.writerow(["weight","fighter1","fighter2","method","round_end","time","notes","event"])
def scrapeWiki(url):
    page = wikipedia.WikipediaPage(url).html()
    soup = BeautifulSoup(page,'html.parser')
    fights = soup.find('table', {'class':'toccolours'}).findAll('tr')
    for fight in fights:
        cells = fight.findAll('td')
        try:
            weight = cells[0].text
            fighter1 = cells[1].text
            fighter2 = cells[3].text
            method = cells[4].text
            roundend = cells[5].text
            time = cells[6].text
            notes = cells[7].text
            writer.writerow([weight,fighter1,fighter2,method,roundend,time,notes,url])
        except:
            pass
for url in urls:
    print("\nTrying: "+url)
    try:
        scrapeWiki(url)
    except:
        print("URL read error: "+url)
```
