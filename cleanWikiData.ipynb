```python
import wikipedia
import csv
from bs4 import BeautifulSoup
import requests
import re
import urllib
import pandas as pd
```

```python
df = pd.read_csv('allUFCfights.csv')
```

```python
df.head()
```

```python
df['time'] = df['time'].str.replace("\n","")
```

```python
df.head()
```

```python
print(re.sub(r'\(.*\)', '', "test (number 3) string (remove this) yes"))
```

```python
df['simple_method']= df['method'].apply(lambda x: re.sub(r'\(.*\)', '', x).strip())
df['simple_method']= df['simple_method'].apply(lambda x: re.sub(r'\[.*\]', '', x))
```

```python
df.head(50)
```

```python
df['simple_method'].value_counts()
```

```python
df['simple_method']= df['simple_method'].apply(lambda x: x.replace("DQ","Disqualification"))
```

```python
for word in ("Unanimous decision", "Split decision", "Unanimous Decision", "Technical Decision"):
    df['simple_method']= df['simple_method'].apply(lambda x: x.replace(word, "Decision"))
```

```python
df['simple_method']= df['simple_method'].apply(lambda x: x.replace("Technical submission", "Technical Submission"))
```

```python
df['simple_method']= df['simple_method'].apply(lambda x: x.replace("Verbal Submission", "Submission"))
```

```python
df['simple_method']= df['simple_method'].apply(lambda x: x.replace("Referee Stoppage", "TKO"))
```

```python
df['simple_method']= df['simple_method'].apply(lambda x: x.replace("Submission", "Submission"))
```

```python
df['simple_weight'].value_counts()
```

```python
df['simple_weight'] = df['weight'].apply(lambda x: re.sub(r'\(.*\)', '', x).strip())
```

```python
df['simple_weight']= df['simple_weight'].apply(lambda x: x.replace("Women’s Featherweight", "Women's Featherweight"))
```

```python
df['simple_weight']= df['simple_weight'].apply(lambda x: x.replace("Women’s Strawweight", "Women's Strawweight"))
```

```python
df['simple_weight']= df['simple_weight'].apply(lambda x: x.replace("Bamtaweight", "Bantamweight"))
```

```python
df['simple_weight']= df['simple_weight'].apply(lambda x: x.replace("Super Heavyweight", "Heavyweight"))
```

```python
df.head()
```

```python
del df['round_end']
del df['time']
```

```python
df['weight'] = df['simple_weight']
df['method'] = df['simple_method']
```

```python
del df['simple_weight']
del df['simple_method']
```

```python
df['class']=df['weight']
del df['weight']
```

```python
df.to_csv('allUFCfightsclean.csv')
```
